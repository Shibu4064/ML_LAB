```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df=pd.DataFrame({
    'x':[12,20,28,18,29,33,24,45,45,52,51,52,55,53,55,61,64,69,72],
    'y':[39,36,30,52,54,46,55,59,63,70,66,63,58,23,14,8,19,7,24]
})

kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
plt.scatter(df['x'],df['y'])
```

```python
labels=kmeans.predict(df)                                           #assigning labels to the data points
centroids=kmeans.cluster_centers_                                          #obtaning the cluster centroids
```

# generating the model and giving Kmeans parameter

```python
fig=plt.figure(figsize=(5,5))                           #figure object with a specified size
colmap={1:'r',2:'g',3:'b'}                                               #define a colormap dictionary
colors=map(lambda x: colmap[x+1],labels) 
colors1=list(colors)                                     #mapping colors based on labels
plt.scatter(df['x'],df['y'],color=colors1,alpha=0.5,edgecolor='k') #plotting the scattered datapoints with assigned colors
for idx,centroid in enumerate(centroids):                 #plotting the centroids which is a list containing the 
    plt.scatter(*centroid, color=colmap[idx+1])           #coordinates of the centroid points
plt.xlim(0,80)                                        #setting x axis limit
plt.ylim(0,80)                                                  #setting y axis limit
plt.show()                                              #displaying the plot
```

# plotting the elbow plot and find K value

```python
sse=[]
k_range=range(1,10)
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['x','y']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.plot(k_range,sse)
```
