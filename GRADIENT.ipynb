```python
import numpy as np
import matplotlib.pyplot as plt
X=np.arange(10)
Y=(X-5)**2
print(X,Y)
```

# given a function f(x), we want to fint the value of x that minimizes f

```python
#visualization
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(X)")
plt.show()
```

```python
x=0
lr=0.1
error=[]
#50 steps in the downhill direction
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
plt.plot(error)
plt.show()
```
